```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data into a DataFrame
data = pd.read_csv('Unemployment_in_India.csv')

# Extract the data from the DataFrame
categories = data['Region']
values = data['Estimated Labour Participation Rate (%)']

# Display the first few rows of the dataset
print(data.head())

# Get summary statistics
print(data.describe())

# Check for missing values
print(data.isnull().sum())
# Handle missing values (if any)
data.dropna(inplace=True)

# Convert date columns to datetime
data['Date'] = pd.to_datetime(data['Date'])
# Plotting a line chart of unemployment over time
plt.figure(figsize=(12, 6))

plt.title('Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate')
plt.grid(True)
plt.show()

# Create a bar graph
plt.bar(categories, values)
# Add labels and a title
plt.xlabel('Category')
plt.ylabel('Values')
plt.title('Comparison of Two Columns from CSV')
# Rotate x-axis labels for better readability if needed
plt.xticks(rotation=95)

# Display the graph
plt.show()

```
